```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.drop('customerID', axis = 1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

df_categorical_clms = [col for col in df.columns if df[col].dtype == 'object']
data_encoded = pd.get_dummies(df, columns = df_categorical_clms)
data_encoded.head()
```

```python
from sklearn.compose import make_column_transformer
```

```python
data_encoded ['Churn'] = df['Churn']
data_encoded['Churn'] = data_encoded['Churn'].map({'Yes':1, 'No':0})
data_encoded = data_encoded.drop(['Churn_Yes', 'Churn_No' ], axis =1)
data_encoded = data_encoded.astype(int)
data_encoded.head()
```

```python
numerical_columns = [col for col in data_encoded.columns if data_encoded[col].dtype == 'int']
scaler = MinMaxScaler()
```

```python
data_encoded[numerical_columns] = scaler.fit_transform(data_encoded[numerical_columns])
data_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from lazypredict.Supervised import LazyClassifier
```

```python
X = data_encoded.iloc[:, :-1]
y = data_encoded.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
lp =LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = lp.fit(X_train, X_test, y_train, y_test)
print(models)
```

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model_name = ['Adaboost CLassifier', 'Random Forest Cassifier', 'LGBM Classifier', 'Logistic Regression']

models = []
models.append(AdaBoostClassifier())
models.append(RandomForestClassifier())
models.append(LGBMClassifier())
models.append(LogisticRegression())
```

```python
for model in models:
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    accuracy = accuracy_score(y_test, preds)
    print(accuracy)
    
```

```python
Model_with_Accurcies = pd.DataFrame({'Models': model_name, 'Accuracies': accuracies})
```
